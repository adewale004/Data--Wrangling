# ACT REPORT

## Introduction
This is a data wrangling project that involves gathering data from a twitter account (@WeRateDogs). WeRateDogs is a twitter account that rates people's dogs with a humorous comment about the dog. Three dataset were gather for this project, which are:
* Enhanced Twitter Archive provided by Udacity in csv format
* Image Prediction dataset used to predict dog breeds gotten programmatically
* Additional twitter data gotten using Twitter API (Tweepy)
Data wrangling process was divided in to three sections, data gathering, assessing data, and cleaning data. A master dataset was created by merging the three dataset together using the tweet ID column. At the end of the data wrangling, the following insights and visualizations were derived from the dataset.

1. What is the distribution of tweet source?
1,638 tweets were made from Iphone, 22 tweets from Twitter Web Client, and 9 tweets from TweetDeck. This shows that 98% of the tweets were from an Iphone device


2. What Dog type is the most reviewed?
Golden retriever (154), Labrador retriever (105), Pembroke (95), Chihuahua (90), and Pug (62) are top 5 most reviewed Dog type respectively.




3. Which Dog type has the most favorite count?
From the visualisation, it can be seen that Saluki, Bedlington terrier, French bulldog, Bouvier_des_Flandres, and Afghan_hound are the Dog type with most likes (favorite count) respectively.



4. What is the relationship between favorite count and retweet count?
There is strong positive correlation  (0.9) between favorite count and retweet count. This mean that as favorite count is increasing, retweet count is also increasing.
